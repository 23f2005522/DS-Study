### Namespaces

A namespace is a space that holds names(identifiers).Programmatically speaking, namespaces are dictionary of identifiers(keys) and their objects(values)

There are 4 types of namespaces:
- Builtin Namespace
- Global Namespace
- Enclosing Namespace
- Local Namespace

### Scope and LEGB Rule

A scope is a textual region of a Python program where a namespace is directly accessible.

The interpreter searches for a name from the inside out, looking in the local, enclosing, global, and finally the built-in scope. If the interpreter doesn’t find the name in any of these locations, then Python raises a NameError exception.

```python
# local and global
# global var
a = 2

def temp():
  # local var
  b = 3
  print(b)

temp()
print(a)
```

```python
# local and global -> same name
## can a same name variable can be present in same code  ? Yes : in different scope / namespace
a = 2

def temp():
  # local var
  a = 3
  print(a)

temp()
print(a)
```

```python
# local and global -> local does not have but global has
a = 2

def temp():
  # local var
  print(a)

temp()
print(a)
```

```python
# local and global -> editing global
## we cant edit global var inside local scope but we can read it
##### as python is not a extremly sctrict language we can edit global var but we need to tell python that we are going to edit global var by using global keyword



a = 2

def temp():
  # local var
  a += 1
  print(a)

temp()
print(a)
```

```python

a = 2
def temp():
  # local var
  global a
  a += 1
  print(a)

temp()
print(a)
```

```python
# local and global -> global created inside local
def temp():
  # local var
  global a
  a = 1
  print(a)

temp()
print(a)
```

```python
# local and global -> function parameter is local
def temp(z):
  # local var
  print(z)

a = 5
temp(5)
print(a)
print(z)
```

```python
# Enclosing scope : if we have a function inside a function then the inner function can access the variables of outer function but outer function cannot access the variables of inner function

def outer():
  def inner():
    print("inner function")
  inner()
  print('outer function')


outer()
print('main program')
```

```python

def outer():
  a = 1
  def inner():
    ## cant 
    a += 1
    print('inner',a)
  inner()
  print('outer',a)


outer()
print('main program')
```

```python
# nonlocal keyword :  it helps to edit the variable of outer function inside inner function 
def outer():
  a = 1
  def inner():
    nonlocal a
    a += 1
    print('inner',a)
  inner()
  print('outer',a)


outer()
print('main program')
```

### Decorators

A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it.

This can happen only because python functions are 1st class citizens.

There are 2 types of decorators available in python
- `Built in decorators` like `@staticmethod`, `@classmethod`, `@abstractmethod` and `@property` etc
- `User defined decorators` that we programmers can create according to our needs

```python
# Python are 1st class function
## first class citizen : are the entities which can be treated as an object and can be passed as an argument to a function and can be returned from a function and can be assigned to a variable

## in python functions are 1st class citizen : 
## exmple of function being a 1st class citizen 

## assigning a function to a variable
          # def square(num):
          #   return num**2
          # s = square
          # print(s(2))


## deleting a function
          # def square(num):
          #   return num**2
          # del square  
          # print(square(2))  :: error as we have deleted the function
          
## passing a function as an argument to another function / higher order function

def modify(func,num):
  return func(num)

def square(num):
  return num**2

modify(square,2)
```

```python
# simple example
# how this works -> closure?  ;; use Python tutorial to understand closure
## clousers : where inner fuction cna access the variables of outer function even after the outer function has finished execution


def my_decorator(f):
  def wrapper():
    print('***********************')
    f()
    print('***********************')
  return wrapper

def hello():
  print('hello')

def display():
  print('hello nitish')

a = my_decorator(hello)
a()

b = my_decorator(display)
b()
```

```python
# Better syntax?
# simple example

def my_decorator(func):
  def wrapper():
    print('***********************')
    func()
    print('***********************')
  return wrapper

@my_decorator
def hello():
  print('hello')

hello()
```

```python
# anything meaningful?
import time

def timer(func):
  def wrapper(*args):
    start = time.time()
    func(*args)
    print('time taken by',func.__name__,time.time()-start,'secs')
  return wrapper

@timer
def hello():
  print('hello wolrd')
  time.sleep(2)

@timer
def square(num):
  time.sleep(1)
  print(num**2)

@timer
def power(a,b):
  print(a**b)

hello()
square(2)
power(2,3)
```

```python
def sanity_check(data_type):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(*args) == data_type:
        func(*args)
      else:
        raise TypeError('Ye datatype nai chalega')
    return inner_wrapper
  return outer_wrapper

@sanity_check(int)
def square(num):
  print(num**2)

@sanity_check(str)
def greet(name):
  print('hello',name)

def RUN(): 
  try : 
    square(2)
  except TypeError as e:
    print("Error:", e)
  finally :
    print('program ended')
    
RUN()
  
def RUN2(): 
  try : 
    square("1")
  except TypeError as e:
    print("Error:", e)
  finally :
    print('program ended')
  
RUN2()
```
