### What is numpy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types


### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

### Changing Datatype


```python
# astype

## generally numpy stores int in 8 bytes (364bits) and float in 8 bytes but we can change the data type of the items in the array using astype method
print(a1, a1.itemsize)
print(a1.astype(np.int32), a1.itemsize)
print(a2.astype(np.str_), a2.itemsize)
print(a2.astype(np.float32), a2.itemsize)
```

### Array Operations


```python
a1 = np.arange(12).reshape(3, 4)
a2 = np.arange(12, 24).reshape(3, 4)

print("a1:")
print(a1)
print("a2:")
print(a2)
```

```python
# scalar operations

# arithmetic : applying arithmetic operations on the array will apply the operation to each item in the array
print(2 * a1)
print(a1**2)
print(a1 + 10)
print(a1 - 10)
print(a1 / 2)
print(a1 // 2)  ## floor division
```

```python
# relational : applying relational operations on the array will apply the operation to each item in the array and return a boolean array

print(a2 == 15)
print(a2 > 15)
print(a2 < 15)
print(a2 >= 15)
print(a2 <= 15)
```

```python
# vector operations ; applying vector operations on the array will apply the operation to each item in the array and ""return"" a new array with the result of the operation
# arithmetic
print("a1 ** a2 = ")
print(a1**a2)
print("matrix Multiplication  ")
print(a1 * a2)
print("matrix Addition : ", a1 + a2)
```

### Array Functions


```python
a1 = np.random.random((3, 3))
a1 = np.round(a1 * 100)
a1
```

```python
# max/min/sum/prod
# 0 -> col and 1 -> row
print(f"Max of all the elements in a1 : {a1.max()}")
print(f"Min of all the elements in a1 : {a1.min()}")
print(f"Sum of all the elements in a1 : {a1.sum()}")
print(f"Product of all the elements in a1 : {np.prod(a1)}")

## row wise and column wise operations
## 0 --> col and 1 --> row

print(f"Max of each column in a1 : {a1.max(axis=0)}")
print(f"Max of each row in a1 : {a1.max(axis=1)})")
print(f"Sum of each column in a1 : {a1.sum(axis=0)}")
print(f"Sum of each row in a1 : {a1.sum(axis=1)}")


np.prod(
    a1, axis=0
)  # .prod() -->  product of elements of the array  either for all elemnts / rows wise each / col wise each
```

```python
# mean/median/std/var
print(f"Mean of all the elements in a1 : {a1.mean()}")
print(f"Median of all the elements in a1 : {np.median(a1)}")
print(f"Standard Deviation of all the elements in a1 : {a1.std()}")
print(f"Variance of all the elements in a1 : {a1.var()}")


## row wise and column wise operations
print(f"Mean of each column in a1 : {a1.mean(axis=0)}")
print(f"Mean of each row in a1 : {a1.mean(axis=1)}")

np.var(a1, axis=1)
```
